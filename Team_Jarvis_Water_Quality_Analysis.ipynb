```
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('cities.csv')  # Replace with your actual file name

# Aggregate data to calculate average Water Pollution by Country
average_data = data.groupby('Country')['WaterPollution'].mean()

# Plot a bar graph for the top 10 countries by average Water Pollution
plt.figure(figsize=(12, 7))
average_data.sort_values(ascending=False).head(10).plot(kind='bar', color='salmon', alpha=0.8)

plt.title('Top 10 Countries by Average Water Pollution', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Water Pollution Index (0-100)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
