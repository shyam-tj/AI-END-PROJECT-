```
import matplotlib.pyplot as plt
import numpy as np

years = [2018, 2019, 2020, 2021]
data = {
    'pH': [[7.0, 7.2, 7.4, 7.6], [7.1, 7.3, 7.5, 7.7], [7.3, 7.5, 7.7, 7.9], [7.4, 7.6, 7.8, 8.0]],
    'Dissolved Oxygen': [[8.0, 8.2, 8.4, 8.6], [7.8, 8.0, 8.3, 8.5], [7.5, 7.7, 7.9, 8.2], [7.6, 7.8, 8.1, 8.4]],
    'Turbidity': [[5.0, 4.8, 4.7, 4.6], [4.5, 4.3, 4.2, 4.1], [3.5, 3.4, 3.3, 3.2], [2.5, 2.4, 2.3, 2.2]],
    'Nitrate Levels': [[10, 10.5, 11, 11.5], [8, 8.5, 9, 9.5], [6, 6.5, 7, 7.5], [12, 12.5, 13, 13.5]],
}

colors = ['purple', 'teal', 'green', 'gold']
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
parameters = list(data.keys())

for i, (param, ax) in enumerate(zip(parameters, axes.flat)):
    for j, values in enumerate(data[param]):
        ax.bar(np.array(years) + (j - 1.5) * 0.2, values, width=0.2, label=['Asia', 'Europe', 'North America', 'South America'][j], color=colors[j])
    ax.set_title(f"{param} Across Regions (2018-2021)")
    ax.set_xlabel("Year")
    ax.set_ylabel(param + (" (mg/L)" if param != "pH" else ""))
    ax.set_xticks(years)
    ax.legend(title="Region")

plt.tight_layout()
plt.show()
```
